```python
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
x = np.arange(10)
y = np.sin(x)
cs = CubicSpline(x, y)
xs = np.arange(-0.5, 9.6, 0.1)
fig, ax = plt.subplots(figsize=(6.5, 4))
ax.plot(x, y, 'o', label='data')
ax.plot(xs, np.sin(xs), label='true')
ax.plot(xs, cs(xs), label="S")
ax.plot(xs, cs(xs, 1), label="S'")
ax.plot(xs, cs(xs, 2), label="S''")
ax.plot(xs, cs(xs, 3), label="S'''")
ax.set_xlim(-0.5, 9.5)
ax.legend(loc='lower left', ncol=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import BSpline, CubicSpline

def spline(x):
    # Define the control points of the B-spline
    points = np.array([[0, 0], [0, 1], [1, 1], [1, 0]])

    # Define the degree of the B-spline (cubic in this example)
    degree = 3

    # Create a BSpline object using the control points and degree
    spline = BSpline(points, degree)

    # Calculate the y-values of the B-spline at the given x-values
    return spline(x)

# Define the knots and boundary conditions
knots = np.asarray([(0,0), (1,1), (2,2), (3,1)])

plt.scatter(*knots.T)

# Use the CubicSpline function to interpolate the spline
cs = CubicSpline(*knots.T)

# Evaluate the spline at the given x-values
x = np.linspace(0, 3, 100)
y = cs(x)

# Plot the spline using matplotlib
plt.plot(x, y)
plt.show()



# # Define the spline as a function of position
# def spline(x):
#   # Calculate the y-values of the spline at the given x-values
#   # This can be done using a polynomial function, a cubic spline,
#   # or any other method for defining a smooth curve
#   y = some_function(x)
#   return y

# Calculate the curvature of the spline at each point along its length
#x = np.linspace(0, 1, 100)  # Sample 100 points along the spline
#y = spline(x)
dy = np.gradient(y)  # First derivative of y with respect to x
d2y = np.gradient(dy)  # Second derivative of y with respect to x
k = d2y / (1 + dy**2)**(3/2)  # Curvature of the spline

# Plot the curvature comb of the spline
plt.plot(x, k)
plt.xlabel("Position along spline")
plt.ylabel("Curvature")
plt.show()
```
